```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['diagnosis'],axis=1)
y=df.diagnosis.values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
print('Test Accuracy: {:.2f}%'.format(lr.score(x_test,y_test)*100))
```

```python
from sklearn.metrics import accuracy_score
y_pred_lr=lr.predict(x_test)
print('Accuracy of model: {:.2f}%'.format(accuracy_score(y_test,y_pred_lr)*100))
```

```python
#SVC
from sklearn.svm import SVC
svc=SVC(random_state=1)
svc.fit(x_train,y_train)
print('Test Accuracy: {:.2f}%'.format(svc.score(x_test,y_test)*100))
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
gb=GaussianNB()
gb.fit(x_train,y_train)
print('Test Accuracy: {:.2f}%'.format(gb.score(x_test,y_test)*100))
```

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100,random_state=5)
rf.fit(x_train,y_train)
print('Test Accuracy:{:.2f}%'.format(rf.score(x_test,y_test)*100))
```

```python
import xgboost
xg=xgboost.XGBClassifier()
xg.fit(x_train,y_train)
print('Test Accuracy: {:.2f}%'.format(xg.score(x_test,y_test)*100))
```
